# BONUS: Put all the previously mentioned data transformations into a function.

```python
cfile = pd.read_csv('marketing_customer_analysis.csv')
cfile
```

```python
def clean_data(df):
    #Standardize column names
    df.columns = [e.lower().replace(' ','_') for e in df.columns]
    
    #drop columns
    df=df.drop(columns = ['unnamed:_0'])
    
    #fill NaN values
    df['vehicles_type'] = df['vehicle_type'].fillna('M')
    
    # chnage dtype of data to Datetime/ create new column 'month'
    
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    df['month'] = df['effective_to_date'].dt.month
    
    #drop remaining NaN values
    df=df.dropna()
    return df
```

```python
cfile2 = clean_data(cfile)
```

# Show a plot of the response rate by the total claim amount.

```python
cfile2['total_claim_amount'].value_counts()
```

```python
print('Maximum value', cfile2['total_claim_amount'].max())
```

```python
print('Minimum value', cfile2['total_claim_amount'].min())
```

```python
print('Mean value', cfile2['total_claim_amount'].mean())
```

```python
def t_bins(x):
    if x <= 450:
        return 1
    elif x <= 900:
        return 2
    elif x <= 1350:
        return 3
    elif x <= 1800:
        return 4
    else:
        return 5
        
cfile2['total_claim_amount_bins'] = list(map(t_bins, cfile2['total_claim_amount']))

cfile2
```

```python
cfile2.groupby('total_claim_amount_bins')['response_rate'].mean().plot.bar()
```

```python
sns.barplot(x= 'total_claim_amount_bins', y='response_rate', data = cfile2)
```

```python
cfile2['total_claim_amount'] = pd.qcut(cfile2['total_claim_amount'], q =15)
sns.barplot(x= 'total_claim_amount', y='response_rate', data = cfile2)
plt.show()
```
